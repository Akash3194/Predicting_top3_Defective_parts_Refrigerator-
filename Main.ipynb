```
from google.colab import files
import io
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
pd.set_option("display.max_row", 150000)
pd.set_option("display.max_columns", 1500)
```

```
uploaded = files.upload()
```

```
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```
refri = pd.read_csv(io.StringIO(uploaded['Refrigerator.csv'].decode('utf-8')))
```

```
def preprocessing(data):
    data.drop(['Unnamed: 0'], axis = 1, inplace = True)
    data.dropna(inplace = True)
    x = data.drop('Part Name', axis = 1)
    x = pd.get_dummies(x)
    
    return x
```

```
x = preprocessing(refri)
y = pd.get_dummies(refri['Part Name'])
y_r = refri['Part Name']
parts = np.array(y.columns)
```

```
x_train, x_test, y_train_r, y_test_r = train_test_split(x, y_r, random_state = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1)
```

```
x_train = np.array(x_train); y_train = np.array(y_train)
x_test = np.array(x_test); y_test = np.array(y_test)

model = keras.Sequential([
    keras.layers.Dense(103, activation = tf.nn.relu),
    keras.layers.Dense(103, activation = tf.nn.relu),
    keras.layers.Dense(103, activation = tf.nn.relu),
 
    keras.layers.Dense(y_train.shape[1], activation = tf.nn.softmax)
])
model.compile(optimizer = tf.train.AdagradOptimizer(0.01),
             loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

```
model.fit(x_train, y_train, epochs= 400)
```

```
k = 1
for i in range(y_test.shape[0]):
  np.set_printoptions(suppress=True)
  top3 = np.argsort(model.predict(x_test[i:i+1,:]))[0][-3:][::-1]
  if y_test_r.iloc[i] in parts[top3]:
      k+= 1
Total_accuracy = k/y_test.shape[0]
```
